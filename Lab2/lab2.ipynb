```python
# Завдання №1
import numpy as np

# Create: Лінійні значення з arange і випадкові числа з random
linear_data = np.arange(10)  # масив від 0 до 9
random_data = np.random.rand(10)  # масив випадкових чисел від 0 до 1

# Read: Зчитування значень масивів
print("Лінійні дані:", linear_data)
print("Випадкові дані:", random_data)

# Update: Оновлення значень
linear_data[0] = 100  # змінюємо перший елемент на 100
print("Оновлені лінійні дані:", linear_data)

# Delete: Видалення елемента
linear_data = np.delete(linear_data, 0)  # видаляємо перший елемент
print("Лінійні дані після видалення:", linear_data)
```

```python
# Завдання №2

import pandas as pd

# Створення DataFrame з масиву NumPy
df_from_numpy = pd.DataFrame({
    'Linear Data': linear_data,
    'Random Data': random_data[:-1]  # Обрізаємо до розміру linear_data
})
print("DataFrame з NumPy масивів:\n", df_from_numpy)

# Створення DataFrame зі словника
data = {'Name': ['Anna', 'Bob', 'Charlie'], 'Age': [23, 30, 35], 'Salary': [5000, 7000, 6500]}
df_from_dict = pd.DataFrame(data)
print("DataFrame зі словника:\n", df_from_dict)

# Операції: head, describe, iloc, loc
print("Перші рядки DataFrame:\n", df_from_dict.head())
print("Опис статистики DataFrame:\n", df_from_dict.describe())
print("Вибір елементу з iloc (перший рядок, всі стовпці):\n", df_from_dict.iloc[0])
print("Вибір елементу з loc (Name='Bob'):\n", df_from_dict.loc[df_from_dict['Name'] == 'Bob'])
```

```python
# Завдання №3

import matplotlib.pyplot as plt

# Згенеруємо лінійну функцію
x = np.linspace(0, 10, 50)
y = 2 * x + 1  # Лінійна функція y = 2x + 1

# Додаємо випадкові похибки
y_with_error = y + np.random.normal(0, 2, size=y.shape)

# Побудова графіка
plt.plot(x, y, label='Лінійна функція')
plt.scatter(x, y_with_error, color='red', label='Дані з похибками')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Лінійна функція та дані з похибками')
plt.show()

# Обчислення MAE та MSE
mae = np.mean(np.abs(y - y_with_error))
mse = np.mean((y - y_with_error) ** 2)
print(f"MAE: {mae:.2f}, MSE: {mse:.2f}")

# Запис результатів у CSV
results = pd.DataFrame({
    'x': x,
    'y_true': y,
    'y_with_error': y_with_error,
    'error': y - y_with_error
})
results.to_csv('results.csv', index=False)
```
