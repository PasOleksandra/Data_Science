 Створити набір даних за допомогою DatasetGenerator

```python
from sklearn.datasets import make_classification
import pandas as pd

# Генерація набору даних з виправленими параметрами
X, y = make_classification(
    n_samples=1000,    # Кількість зразків
    n_features=20,     # Кількість ознак
    n_informative=2,   # Кількість інформативних ознак
    n_redundant=10,    # Кількість надлишкових ознак
    n_classes=2,       # Кількість класів (зменшено до 2)
    n_clusters_per_class=2,  # Кількість кластерів на клас
    random_state=42    # Фіксований випадковий стан для відтворюваності
)

# Перетворення в DataFrame
df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])

# Додавання міток класів
df['target'] = y

# Збереження в CSV файл
df.to_csv('generated_dataset.csv', index=False)

print("Дані збережено в файл 'generated_dataset.csv'.")
```

(Опціонально для отримання відмінної оцінки) Згенерувати набір даних для класифікації, використовуючи стандартні можливості Python (наприклад, з допомогою бібліотек Numpy або random)

```python
import numpy as np
import pandas as pd
import random

# Налаштування параметрів
n_samples = 1000  # Кількість зразків
n_features = 20   # Кількість ознак
n_classes = 3     # Кількість класів
random_state = 42 # Для відтворюваності результатів

# Встановлення випадкових чисел
np.random.seed(random_state)
random.seed(random_state)

# Генерація випадкових ознак
X = np.random.randn(n_samples, n_features)  # Нормальний розподіл для ознак

# Генерація випадкових міток класів
y = np.random.randint(0, n_classes, n_samples)  # Випадкові цілі для класів (0, 1, 2)

# Створення DataFrame
df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(n_features)])
df['target'] = y

# Збереження в CSV файл
df.to_csv('generated_classification_data.csv', index=False)

print("Дані збережено в файл 'generated_classification_data.csv'.")
```

Моделювання власного набору даних

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Завантаження згенерованого набору даних (замість '/path/to/data.csv' вказати ваш шлях)
data = pd.read_csv('generated_classification_data.csv')

# Поділ на вхідні змінні (X) та цільову змінну (y)
X = data.drop('target', axis=1)  # Ознаки
y = data['target']  # Цільова змінна (мітки класів)

# Поділ даних на тренувальний і тестовий набори
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Створення та тренування моделі логістичної регресії
model = LogisticRegression(max_iter=1000)  # Параметр max_iter для збільшення кількості ітерацій
model.fit(X_train, y_train)

# Прогнозування на тестових даних
y_pred = model.predict(X_test)

# Оцінка якості моделі
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Класифікаційний звіт
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Матриця плутанини
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Візуалізація результатів (для двовимірного набору даних, якщо X має більше двох ознак, то буде складно зробити графік)
plt.figure(figsize=(8, 6))
plt.scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=y_pred, cmap='coolwarm', edgecolors='k', s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Classification Results (Logistic Regression)')
plt.show()
```

Моделювання завантажених даних

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Завантаження даних
file_path = 'cyberbullying_tweets.csv'
data = pd.read_csv(file_path)

# Перегляд перших кількох рядків даних
print(data.head())

# Перевірка на наявність пропущених значень
print(data.isnull().sum())

# Розподіл на вхідні змінні (X) та цільову змінну (y)
X = data['tweet_text']  # Тексти твітів
y = data['cyberbullying_type']  # Мітки (0 - не кібербулінг, 1 - кібербулінг)

# Перетворення тексту в числові ознаки за допомогою TF-IDF векторизації
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_vec = vectorizer.fit_transform(X)

# Розбиття на тренувальний і тестовий набори (80% для тренування, 20% для тестування)
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

# Створення моделі логістичної регресії
model = LogisticRegression(max_iter=1000)

# Тренування моделі
model.fit(X_train, y_train)

# Прогнозування на тестових даних
y_pred = model.predict(X_test)

# Оцінка якості моделі
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Класифікаційний звіт
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Матриця плутанини
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Візуалізація матриці плутанини
plt.figure(figsize=(7, 5))
plt.imshow(confusion_matrix(y_test, y_pred), interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = range(2)
plt.xticks(tick_marks, ['Non-cyberbullying', 'Cyberbullying'])
plt.yticks(tick_marks, ['Non-cyberbullying', 'Cyberbullying'])
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()

```
